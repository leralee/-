```python
import networkx as nx
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
G = nx.DiGraph()
```

```python
E = [('v0', 'v1'), 
      ('v1', 'v2'),
     ('v2','v3'), 
      ('v3', 'v4'),
     ('v4', 'vT'),
      ('v0', 'v5'), 
     ('v5', 'v6'), 
    ('v6', 'v7'), 
     ('v7', 'v11'), 
     ('v11', 'vT'),
    ('v7', 'v4'),
       ('v0', 'v8'),
     ('v8', 'v9'),
       ('v9', 'v10'),
     ('v10', 'v11'),
       ('v6', 'v10'),
    ('v3', 'v11'),
    ('v8', 'v6')]
```

```python
G.add_edges_from(E)
```

```python
attrs = {('v0', 'v1'): {'time': 2,'cost': 5, 'weights': (2,5)}, 
      ('v1', 'v2'): {'time': 2,'cost': 4, 'weights': (2,5)},
     ('v2','v3'): {'time': 4,'cost': 1, 'weights': (4,1)}, 
      ('v3', 'v4'):{'time': 2,'cost': 0, 'weights': (2,0)},
     ('v4', 'vT'):{'time': 1,'cost': 3, 'weights': (1,3)},
      ('v0', 'v5'): {'time': 2,'cost': 1, 'weights': (2,1)}, 
     ('v5', 'v6'): {'time': 2,'cost': 2, 'weights': (2,2)}, 
    ('v6', 'v7'): {'time': 1,'cost': 1, 'weights': (1,1)}, 
         ('v7', 'v11'): {'time': 1,'cost': 1, 'weights': (1,1)}, 
          ('v11', 'vT'):{'time': 2,'cost': 1, 'weights': (2,1)} ,
    ('v7', 'v4'):{'time': 2,'cost': 3, 'weights': (2,3)},
       ('v0', 'v8'): {'time': 3,'cost': 5, 'weights': (3,5)},
     ('v8', 'v9'): {'time': 4,'cost': 3, 'weights': (4,3)},
       ('v9', 'v10'):{'time': 2,'cost': 3, 'weights': (2,3)},
     ('v10', 'v11'):{'time': 4,'cost': 1, 'weights': (4,1)},
       ('v6', 'v10'):{'time': 2,'cost': 2, 'weights': (2,2)},
    ('v3', 'v11'): {'time': 1,'cost': 1, 'weights': (1,1)},
        ('v8', 'v6'): {'time': 3, 'cost': 3, 'weights': (3,3)}
        }
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FixedLocator

x = list_of_time
y = list_of_cost

fig = plt.figure(figsize=(7,7), dpi=300)
axes = fig.add_axes([.05,.05,.9,.9])

# Plots the data
axes.scatter(x, y, color='green')
axes.vlines(x, 0, y, linestyle="dashed", color='green')
axes.hlines(y, 0, x, linestyle="dashed", color='green')

axes.set_xlim(0, 17)
axes.set_ylim(0, 17)
axes.set_xticks(x)
axes.set_yticks(y)
axes.invert_yaxis()

# Move ticks to top side of plot
axes.xaxis.set_tick_params(
    length=0, bottom=False, labelbottom=False, top=True, labeltop=True
)
axes.xaxis.set_tick_params(length=0)
axes.yaxis.set_tick_params(length=0)

# Add arrows to the spines by drawing triangle shaped points over them
axes.plot(1, 1, '>k', transform=axes.transAxes, clip_on=False)
axes.plot(0, 0, 'vk', transform=axes.transAxes, clip_on=False)
axes.spines[['bottom', 'right']].set_visible(False)

# Add labels for 0, F_1 and F_2
from matplotlib.transforms import offset_copy
axes.text(
    0, 1, s='0', fontstyle='italic', ha='right', va='bottom',
    transform=offset_copy(axes.transAxes, x=-5, y=5, fig=fig, units='points'),
)
axes.text(
    1, 1, s='$F_1$', fontstyle='italic', ha='right', va='bottom',
    transform=offset_copy(axes.transAxes, x=0, y=5, fig=fig, units='points'),
)
axes.text(
    0, 0, s='$F_2$', fontstyle='italic', ha='right',
    transform=offset_copy(axes.transAxes, x=-5, y=0, fig=fig, units='points'),
)

# Add labels at each point. Leveraging the alignment of the text
# AND padded offset.
lc = ('top', 'center', 0, -5)
ll = ('top', 'right', -5, -5)
lr = ('top', 'left', 5, -5)
ur = ('bottom', 'left', 5, 5)
alignments = [lc, lr, ll, ll, lc, ll, ll, ur, lc]
for i, (xc, yc, (va, ha, padx, pady)) in enumerate(zip(x, y, alignments), start=1):
    axes.annotate(
        xy=(xc, yc), xytext=(padx, pady),
        text=f'$F(x_{i})$', ha=ha, va=va, textcoords='offset points')

plt.show()
```

```python
copy_G = G.copy()
time = []
cost = []


for i in range(10):
    random_edges = random.sample(list(copy_G.edges), 3)
    for u, v in random_edges:
        copy_G[u][v]['time'] = random.randint(1,4)
        copy_G[u][v]['cost'] = random.randint(1,4)
        
        
    simple_path = list(nx.all_simple_paths(copy_G, 'v0', 'vT'))

    
    local_list_of_time = []
    local_list_of_cost = []
    

    
    for j in simple_path:
        local_list_of_time.append(nx.path_weight(copy_G, j, weight='time'))
    
    for k in simple_path:
        local_list_of_cost.append(nx.path_weight(copy_G, k, weight='cost'))
        
    min_value_tuple = min(list(zip(local_list_of_time, local_list_of_cost)))
    
    time.append(min_value_tuple[0])
    cost.append(min_value_tuple[1])
    
    
        
    ##   ,      
    
    
        
#     T = nx.minimum_spanning_tree(copy_G)
#     edge_T = T.edges
#     spl = dict(nx.all_pairs_dijkstra_path_length(copy_G))
#     summ = 0
#     for edge in copy_G.edges:
#         if edge in edge_T:
#             summ+=(spl[edge[0]][edge[1]])
#     x_cor.append(i)
#     y_cor.append(summ)

```
